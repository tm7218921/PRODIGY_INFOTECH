```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load and clean data
year = "2023"
data_2023 = data[["Country Name", year]].dropna()
data_2023[year] = pd.to_numeric(data_2023[year], errors="coerce")

# Select top 10 countries by population for bar chart
top_countries = data_2023.sort_values(by=year, ascending=False).head(10)

# Create bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=top_countries, x="Country Name", y=year, palette="viridis")
plt.title("Top 10 Countries by Population (2023)", fontsize=16)
plt.xlabel("Country", fontsize=12)
plt.ylabel("Population", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random

# Simulated dataset for age and gender
population_size = 1000

# Generate age data (0 to 100 years)
ages = [random.randint(0, 100) for _ in range(population_size)]

# Generate gender data (Male or Female)
genders = [random.choice(["Male", "Female"]) for _ in range(population_size)]

# Create a DataFrame
data = pd.DataFrame({"Age": ages, "Gender": genders})
```

```python
# Bar chart for gender distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x="Gender", palette="pastel")

plt.title("Gender Distribution in Population", fontsize=16)
plt.xlabel("Gender", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.show()
```

```python
# Histogram for age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data["Age"], bins=20, kde=True, color="lightblue")

plt.title("Age Distribution in Population", fontsize=16)
plt.xlabel("Age", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
